```python
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

```python
df = pd.read_csv('../time-series/data/incoming_packages.csv')
```

```python
df.head()
```

```python
df = df[["date", "total_packages"]]
```

```python
df.head()
```

```python
df["moving_10"] = df["total_packages"].rolling(10).mean()
```

```python
df.head(20)
```

```python
df["moving_20"] = df["total_packages"].rolling(20).mean()
```

```python
df["moving_50"] = df["total_packages"].rolling(50).mean()
```

```python
df["moving_100"] = df["total_packages"].rolling(100).mean()
```

```python
plt.plot(df['date'], df['total_packages'])
```

```python
df = df[df["date"] >= "2021-06-01"]
```

```python
df.head()
```

```python
plt.plot(df["date"], df["total_packages"])
```

```python
plt.plot(df['date'], df['total_packages'], label="Normal")
plt.plot(df['date'], df['moving_10'], label="Moving 10")
plt.plot(df['date'], df['moving_20'], label="Moving 20")
plt.plot(df['date'], df['moving_50'], label="Moving 50")
plt.plot(df['date'], df['moving_100'], label="Moving 100")
plt.xticks([
    "2021-06-01",
    "2021-07-01",
    "2021-08-01",
    "2021-09-01",
    "2021-10-01",
    "2021-11-01",
    "2021-12-01"
])
plt.legend(loc="upper right")
plt.show()
```
